### Calzetti (2012) - Star formation rate indicators - Lecture note (02/27/2019)

https://arxiv.org/abs/1208.2997

- Stellar populations mix within galaxies on timescales that are comparable to those of their UV light lifetime. The stellar IMF (i.e., the distribution of stellar masses at birth) may or may not be fully sampled locally. The star formation history may vary from region to region. Both young and old stars can heat the dust in a galaxy, and the dust spectral energy distribution (SED) and features provide little discrimination as to the source of the heating. Because of all these reasons, local SFR indicators are by far less settled than the global ones.

- the basic goal is to identify emission that probes newly or recently formed stars, while avoiding as much as possible contributions from evolved stellar populations

- Most researchers would agree that ‘recent’ refers to timescales ≈10–100Myr when considering whole galaxies, and ≈1–10Myr when considering regions or structures within galaxies (e.g., giant molecular clouds, etc.).


- YSO counting

$${\rm SFR}({\rm YSO}) = N_{\rm YSO}\dfrac{\langle M\rangle}{\tau}$$


- For unresolved systems need assumptions about

    - stellar IMF is known and fully sampled
    - constant SFR over the timescale probed by the specific emission being used
    
    
    - UV: direct starlight
    - IR: starlight reprocessed by dust
    - Ionizing photon rate: gas ionized by massive stars (photoionization > shock ionization). Recombination lines (optical through near-IR). Forbidden metal lines, free-free.
    - HMXB, massive stars, SNe.
    - Synchrotron emission (Condon 1992): cosmic rays are produced and accelerated in supernova remnants, and core-collpase represent 70% or more of the total SNe in star-forming galaxies
    
- 1.2.1.1 Indicators based on direct stellar light: 
    - Lifetime of O6 star ~ 6 Myr
    - Lifetime of B8 star ~ 350 Myr
    - __Luminosity ratio at $0.16\mu {\rm m}$ of an O6 to a B8 star is $\sim 90$. For every O6 star formed, about 150 B8 stars are formed. Thus, at age zero, the UV emission from the collective contribution of B8 stars is comparable to that of O6 stars.__

### Conroy & Kratter (2012) - Runaway stars and the escape of ionizing radiation from high-redshift galaxies (02/13/2019)

http://adsabs.harvard.edu/abs/2012ApJ...755..123C

* Galaxies in high-z galaxies were smaller (say, $\lesssim 100{\rm pc}$) and runaways would be able to travel far from the inner regions of their host galaxies.

* Do slow runaways tend to have short lifetimes?

##### Introduction

* Escape fraction in the range of a few percent for normal star-forming galaxies in the local universe (Leitherer+95; Bland-Hawthorn+99; Heckman+01; Grimes+07)

* For high-z, massive galaxies, $f_{\rm esc}$ increases to $\sim 10\%$.

* Differences between model predictions can be attributed to difference choices for physical parameters and numerical models as well as subgrid models.

* Clarke & Oey (2002): critical SFR above which the porosity of the ISM is approximately unity.

* Fujita et al. (2003): High porosity - high escape fraction

* Runaway stars: Blaauw 1961; Gies & Bolton 1986. Two formation channels: dynamical ejection from young dense stellar systems. Binary origin - explosion of a companion star. Or combination of two (Hoogerwerf+01).

* __What is the fraction of massive stars that are runaways?__: 0.1-0.5 depending on definition and correction for completeness and observational biases.


* __Potential importance of runaways first suggested by Dove \& Shull (1994).__

#### Model
Gas scale radius $r_g = j_d\lambda/(\sqrt{2}f_b)r_{\rm vir}$. Take $j_d=f_b$ and $\lambda \sim 0.05$.


* Probability of a star with ionizing luminosity $Q$? - Need stellar IMF, MS lifetime-mass relation, mass-ionizing-luminosity relation, and a star cluster mass function. Or more simply, the luminosity function of $Q$ with power-law index -2.0.

* __Essentially nothing is known about the dust composition or grain size distribution in high-z galaxies.__

* To account for the effect of dust absorption, they attnuate the H ionizing luminosity Q by a factor $e^{-\tau_{\rm d}}$, where $\tau_{\rm d} = \sigma_{\rm d}N_{\rm H}$. <font color='red'>__This is wrong!!__</font>

##### Runaways

* $\color{red}{\text{Runaways produced by dynamical ejection acquire high velocity not long after their birth.}}$

* Dynamical runaways may not represent the overall massive star population because dynamical encounters tend to eject the lowest mass star in the encounter (Anosova 1986).


* Runaways created by supernova mechanism: lifetime of runaway shorter because they are formed after the death of the primary.

* The luminosity function of supernovae runaway determined by a combination of the IMF, MS lifetime, and binary properties.


* __Dynamically ejected runaways have much large impact on f_esc.__

##### Results

* Model dependencies
    * galaxy size relative to the distance travelled by runaways
    * disk or sphere - runaways travelling perpendicular to the disk
    * velocity of runaways
    * dust absorption
    * Reducing maximum stellar mass
    
##### Caveats

* Lack of any detailed model for ISM structure shaped by stellar feedback, radiative transfer.

    - Flux
$$ {\rm Flux} = S = \int 4\pi j' dM/(4\pi D^2),$$

    $$ j = \rho \kappa B(T_{\rm dust}) = \alpha B(T_{\rm dust})$$

### Kreckel+2018 - A 50 pc Scale View of Star Formation Efficiency across NGC 628 (02/06/2019)

http://adsabs.harvard.edu/abs/2018ApJ...863L..21K

- ALMA observation of CO(2-1) line emission at 1’’ resolution - 47pc
- Median cloud mass - 5x10^5 Msun
- MUSE observations. Extinction correction using Balmer decrement, assuming T_i = 10^4 K and R_V=3.1.
- Catalog of HII regions using HIIPhot: mean luminosity - 10^37 erg/s
- Breakdown of K-S relation on scales smaller than 1kpc due to stochastic sampling - displacement of CO and Halpha
- Significant regions detected in only one tracers
- Increased scatter in depletion time on smaller scales

“If neighboring star-forming regions are independent and uncorrelated, with evolutionary cycling as the only source of scatter, then we naively expect this relation to follow a simple power law (σ ∝ R−1; Leroy et al. 2013). However, accounting for additional scatter from the cloud or HII region mass spectrum, flux evolution within a single evolutionary phase, and sensitivity limits, we expect uncorrelated SF to result in a flattening at the smallest spatial scales (Kruijssen & Longmore 2014; Kruijssen et al. 2018).”

- Out of 738 GMCs and 1502 HII regions only 74 match - implying a rapid disruption by feedback.
